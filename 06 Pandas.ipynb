```python
# Series
Non homogeneous (Like a list)
Custom Index
```

```python
import pandas as pd

ser1 = pd.Series([10,'A',30], index=[1,2,3])
print(ser1)
```

```python
# As dictionary. Keys are index
pd.Series({'A':10, 'B':20})
```

```python
print(ser1[3])
print(ser1[1:3])
```

```python
# Arithmetic respects index
ser1 = pd.Series({'a':10, 'b':20, 'c':30})
ser2 = pd.Series({'a':100, 'b':200, 'd':'This is string'})
ser1+ser2
```

```python
# NaN- not a number (missing value)
```

```python
# Alias to class name
Ser = pd.Series
ser3 = Ser({'a':10, 'b':20, 'c':30})
ser3
```

```python
# Append doesn't respect index
ser1.append(ser2)
```

```python
# Select multiple using a list
ser1[['a', 'b']]
```

```python
ser1.size
ser1.index
```

```python
# DataFrame
# Create as dictionary of series (each series as a column)
df = pd.DataFrame({'col1' : ser1, 'col2' : ser2})
print(df)
```

```python
# Create as list of tuples (each tuple as a row)
pd.DataFrame([(1,2,3), (10,20,30)])
```

```python
# Indexing variations
# By columns
df['col1']
df[['col1', 'col2']]
df.col1
# No slicing
```

```python
# By row
df[1:3]  # Original index, not custom index
df.loc['a']  # Custom index
df.loc[['a', 'b']]

# Access cell
df.loc['a', 'col1']

# By original index (not custom index)
df.iloc[0, 0]
```

```python
print(df.size)
print(df.shape)
print(df.index)
print(df.columns)
```

```python
# Get a glance into the data
df.info()
df.describe()
df.head()
```

```python
# Add or edit columns
df.assign(col3 = pd.Series({'a':1000,'b':2000,'c':3000}))
```

```python
# Many functions to read from files
pd.read_csv()
pd.read_excel()  # Requires xlrd library installed
...
```

```python
df = pd.read_excel('USX ways May2007.xls')
```

```python
# Specify which row contains column names and which is to act as the row index.
df = pd.read_excel('USX ways May2007.xls', header=1, index_col=0)
```

```python
# Many functions to write to files. Requires intermediate wirter objects.
df.to_csv()
df.to_excel(pd.ExcelWriter)  # Requires xlwd library installed
...
```

```python
# Add new column
df = df.assign(NewCol = pd.Series(range(100,130), index=range(1,31)))
```

```python
# Display columns in different sequence.
# Internal dictionary doesn't care about order of columns.
df[['NewCol', 'Percent']]
```

```python
# Write to file
# Normal way not working. Requires use of with-as statement
writer = pd.ExcelWriter('new.xls')
df.to_excel(writer)
```

```python
# Works
with pd.ExcelWriter('.\\new.xls') as writer:
    df.to_excel(writer)
```
